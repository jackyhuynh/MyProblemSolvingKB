```python
import random
from IPython.display import clear_output

logo = """
.------.            _     _            _    _            _    
|A_  _ |.          | |   | |          | |  (_)          | |   
|( \/ ).-----.     | |__ | | __ _  ___| | ___  __ _  ___| | __
| \  /|K /\  |     | '_ \| |/ _` |/ __| |/ / |/ _` |/ __| |/ /
|  \/ | /  \ |     | |_) | | (_| | (__|   <| | (_| | (__|   < 
`-----| \  / |     |_.__/|_|\__,_|\___|_|\_\ |\__,_|\___|_|\_\\
      |  \/ K|                            _/ |                
      `------'                           |__/           
"""


"""
Retun a random card from the deck
"""
def dealCard():
    cards=[11,2,3,4,5,6,7,8,9,10,10,10,10]
    return random.choice(cards)

"""
Calculate the score: take a list of cards return the score return from the cards
"""
def calculateScore(cards):
    if sum(cards)== 21 and len(cards)==2:
        return 0
    if 11 in cards and sum(cards)>21:
        cards.remove(11)
        cards.append(1)
    return sum(cards)

def compare(userScore, pcScore):
    if userScore==pcScore:
        return "Draw :)"
    elif pcScore==0:
        return "User Lose, PC black jack"
    elif userScore == 0: 
        return "You Win"
    elif userScore>21:
        return "Over 21! You lose"
    elif pcScore > 21:
        return "You win! PC over 21"
    elif userScore>pcScore:
        return "You win!!"
    else:
        return "You lose!!"
    
def playGame():
    print(logo)
    
    # Hold user and computer cards
    userCards=[]
    pcCards=[]    
    isGameOver=False

    # Deal Cards for both:
    for _ in range(2):
        userCards.append(dealCard())
        pcCards.append(dealCard())
    
    # User pick cards
    while not isGameOver:
        userScore=calculateScore(userCards)
        pcScore=calculateScore(pcCards)
        print(f"User score: {userScore} of {userCards}")
        print(f"Computer's first cards: {pcCards[0]}")

        if userScore == 0 or pcScore == 0 or userScore>21:
            isGameOver=True
        else: 
            if input(f"If user want another card, type 'y', or 'n' to pass")=='y':
                userCards.append(dealCard())
            else:
                isGameOver=True

    while pcScore != 0 and pcScore < 17:
        pcCards.append(dealCard())
        pcScore=calculateScore(pcCards)

    print(f"Your final hands: {userCards}, final score: {userScore}")
    print(f"Computer score: {pcCards}, pc score: {pcScore}")
    print(compare(userScore, pcScore))
    
while input("Do you want to play a game of Blackjack? Type 'y' or 'n': ") == "y":
    clear_output()
    playGame()
```
