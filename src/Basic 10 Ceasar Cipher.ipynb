```python
logo = """           
 ,adPPYba, ,adPPYYba,  ,adPPYba, ,adPPYba, ,adPPYYba, 8b,dPPYba,  
a8"     "" ""     `Y8 a8P_____88 I8[    "" ""     `Y8 88P'   "Y8  
8b         ,adPPPPP88 8PP"""""""  `"Y8ba,  ,adPPPPP88 88          
"8a,   ,aa 88,    ,88 "8b,   ,aa aa    ]8I 88,    ,88 88          
 `"Ybbd8"' `"8bbdP"Y8  `"Ybbd8"' `"YbbdP"' `"8bbdP"Y8 88   
            88             88                                 
           ""             88                                 
                          88                                 
 ,adPPYba, 88 8b,dPPYba,  88,dPPYba,   ,adPPYba, 8b,dPPYba,  
a8"     "" 88 88P'    "8a 88P'    "8a a8P_____88 88P'   "Y8  
8b         88 88       d8 88       88 8PP""""""" 88          
"8a,   ,aa 88 88b,   ,a8" 88       88 "8b,   ,aa 88          
 `"Ybbd8"' 88 88`YbbdP"'  88       88  `"Ybbd8"' 88          
              88                                             
              88           
"""

alphabet = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z']



#TODO-1: Create a function called 'encrypt' that takes the 'text' and 'shift' as inputs.
def encrypt(input_text, shift_amount):
    encryptText=''
    for letter in input_text:
        index = alphabet.index(letter)
        position = (index + shift_amount)%26
        encryptText += alphabet[position]
    print(f"The encode is {encryptText}")
        
    
def decrypt(cipher_text, shift_amount):
    decryptText=''
    for letter in cipher_text:
        index = alphabet.index(letter)
        position = ((index - shift_amount))%26
        decryptText+=alphabet[position]
    print(f"The encode is {decryptText}")    

def cipher(input_text, shift_amount, cipher_direction):
    endtext=""
    if cipher_direction=='decode':
        shift_amount*=-1
    for char in input_text:
        if char in alphabet:
            index= alphabet.index(char)
            position= (index + shift_amount)%26
            endtext+=alphabet[position]
        else:
            endtext+=char
    print(f"Here's the {cipher_direction}d result: {endtext}")
    
```
